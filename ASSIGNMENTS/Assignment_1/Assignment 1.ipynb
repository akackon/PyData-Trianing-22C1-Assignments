```python
'''
ASSIGNMENT 1
You are to scrape the Product Title and Price from https://www.jumia.com.gh/laptops/
'''
# Because Jumia's website is protected by Cloudflare, I am using the clou
import cloudscraper
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd

# instantiate the cloudscraper
rq = cloudscraper.create_scraper()
# place in the url of the site to be scraped
url = "https://www.jumia.com.gh/laptops/"
# perform a get request
res = rq.get(url)

# parse the request using beautiful soup
laptops = bs(res.content)

# find the exact content on the website you are to scrape.
c_name_results = laptops.find_all("h3",class_="name")
c_price_results = laptops.find_all("div", class_="prc")

# extract exact content you scraped
laptop_names = [str(t)[17:-5] for t in c_name_results]
laptop_price = [t.text.strip() for t in c_price_results]
# len(laptop_names)
# len(laptop_price)

# setup data frames

d1 = {"Laptop Names":laptop_names[:40],"Laptop Price":laptop_price[:40]}
df = pd.DataFrame(d1)

df.to_csv("assignment1.csv")

df
```
